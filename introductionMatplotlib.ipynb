```python
import matplotlib.pyplot as plt # 导入图形库
%matplotlib inline 
# 显示，避免用命令 plt.show()
import numpy as np
```

```python
x = np.linspace(0,5,40)
y = np.sin(2*x)
y1 = np.zeros(len(x))

plt.plot(x,y,'r-s',linewidth=1,markersize=2);
plt.plot(x,y1,'b-');
plt.xlabel('x');
plt.ylabel('y');
plt.title("title");
#plt.show()
```

```python
x = np.linspace(0,10,20)
y1 = x**2
y2 = x**1.5

plt.plot(x,y1,"gs--",markersize=6,linewidth=1,markerfacecolor="None",label="first")
plt.plot(x,y2,"ro--",markersize=6,linewidth=1,markerfacecolor="None",label="second")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("Function Pics")
plt.axis([0.5,10,0.2,100]) # plt.axis([Xmin,Xmax,Ymin,Ymax])
plt.legend(loc="best")
plt.savefig("myplot.pdf")
```

- `semilogx()`
- `semilogy()`
- `loglog()`

```python
x = np.logspace(-1,1,20) # logspace based on 10
y1 = x**2
y2 = x**1.5
plt.loglog(x,y1,"gs--",markersize=6,linewidth=1,markerfacecolor="None",label="first")
plt.loglog(x,y2,"ro--",markersize=6,linewidth=1,markerfacecolor="None",label="first")
```

```python
x = np.linspace(0,5,100)
y = np.exp(x)

plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.title("linear axis")
plt.subplot(1,2,2)
plt.semilogy(x,y)
plt.xlabel("$X$")
# plt.ylabel("$Y$")
plt.title("log y axis");
```

# Histograms 

```python
x = np.random.normal(size=1000)
```

```python
plt.hist(x);
```

```python
plt.hist(x,density=True,bins=np.linspace(-5,5,21)); # 21 points generating 20 bins
```

```python
x = np.random.gamma(2,3,100000)
```

```python
plt.hist(x,bins=400,density=True,histtype="step");
plt.grid()
```

```python
plt.hist(x,bins=40,cumulative=True,density=True,histtype="step");
```

```python
x = np.random.gamma(2,3,100000)
plt.figure()
plt.subplot(221)
plt.hist(x,bins=200);
plt.subplot(222)
plt.hist(x,bins=200,density=True);
plt.subplot(223)
plt.hist(x,bins=200,density=True,cumulative=True);
plt.subplot(224)
plt.hist(x,bins=200,density=True,histtype="step");
plt.tight_layout()
```

```python
x = np.linspace(0,7,40)
y = np.sin(x)
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
x = np.linspace(0,7,40)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x)
y4 = np.tan(x)

fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0,0].plot(x,y1)
axes[0,1].plot(x,y2)
axes[1,0].plot(x,y3)
axes[1,1].plot(x,y4)

plt.tight_layout()
```
