# 第一题

计算傅里叶级数：$$\frac{\pi^2}{3}+4\sum^{\infty}_{n=1}(-1)^n\frac{\cos nx}{n^2}$$ 并用图像显示该级数在区间 $(- \pi,\pi)$ 收敛于 $x^2$。

```python
import matplotlib.pyplot as plt # 导入图形库
import math
import numpy as np
%matplotlib inline
```

```python
def function(x,N):
    y=((np.pi)*(np.pi)/3)
    for i in range(1,N,1):
        y=y+4*pow(-1,i)*np.cos(i*x)/(i*i)
    return (y)      
```

```python
x = np.arange(-np.pi,np.pi,0.01)
N = np.arange(10,50,10)
y = np.zeros([np.size(x,0),np.size(N,0)])
m = 1
for j in N:
        y[:,m-1] = function(x,j)
        m = m+1
plt.figure(figsize=(16, 8))
for i in np.arange(np.size(x,0)):
    plt.plot(x,y[:,0],x,y[:,1],x,y[:,2],x,y[:,3],x,x*x,'r')
plt.legend(['N=10','N=20','N=30','N=40','x^2'])
plt.title('$y-x$')
plt.xlabel("$x$")
plt.ylabel("$y$")
```

# 第二题

标准映射(*Standard Map*)是著名的保守二维混动映射，表达式如下：
$$ \left\{
\begin{array}{ll}
p_{n+1}  = p_n + k \sin x_n \\
x_{n+1}  =x_n + p_{n+1}   \\
\end{array}
\right.
$$

$k$ 是控制参数，每迭代一次后，$p_{n+1}$ 和 $x_{n+1}$ 都取 $2 \pi $ 的模 (`%(2*np.pi)`)，这样，系统的相空间就约束在 $[0,2 \pi] \times [0,2 \pi]$ 的平面上。从相空间中随机挑选**若干点**作为初始条件同时启动迭代(也就是同时演化多条轨道)。画出 $k=0.6$ 和 $k=0.9$ 的相图 ($p$ vs $x$)。(注意：$k$=0 时是一个平庸解，可以用来检验程序的正确性。)

```python
import matplotlib.pyplot as plt # 导入图形库
import math
import numpy as np
%matplotlib inline
```

```python
#k=0.6的定义函数解法
plt.figure(figsize=(10, 10))
def function():
    x=[np.random.uniform()]
    p=[np.random.uniform()]
    for n in range(0,1000,1):
        y=p[n]+0.6*(np.sin(x[n]))
        y2=y%(2*np.pi)
        b=x[n]+y
        b2=b%(2*np.pi)
        p.append(y2)
        x.append(b2)
    plt.plot(p,x,'.')#画k=0.6的相图
    plt.title('$k=0.6$')
    plt.xlabel("$p$")
    plt.ylabel("$x$")
for i in range(10):
    function()
```

```python
#k=0.9的定义函数解法
plt.figure(figsize=(10, 10))
def function():
    x=[np.random.uniform()]
    p=[np.random.uniform()]
    for n in range(0,1000,1):
        y=p[n]+0.9*(np.sin(x[n]))
        y2=y%(2*np.pi)
        b=x[n]+y
        b2=b%(2*np.pi)
        p.append(y2)
        x.append(b2)
    plt.plot(p,x,'.')#画k=0.9的相图
    plt.title('$k=0.9$')
    plt.xlabel("$p$")
    plt.ylabel("$x$")
for i in range(10):
    function()
```

```python
#k=0的定义函数解法
plt.figure(figsize=(10, 10))
def function():
    x=[np.random.uniform()]
    p=[np.random.uniform()]
    for n in range(0,1000,1):
        y=p[n]+0*(np.sin(x[n]))
        y2=y%(2*np.pi)
        b=x[n]+y
        b2=b%(2*np.pi)
        p.append(y2)
        x.append(b2)
    plt.plot(p,x,'.')#画k=0的相图
    plt.title('$k=0$')
    plt.xlabel("$p$")
    plt.ylabel("$x$")
for i in range(10):
    function()
```

# 第四题

Duffing 振子是带阻尼的受迫振动，其运动方程为：$$\ddot{x}+\gamma \dot{x}-\omega^2x+\epsilon x^3 = \Gamma cos(\Omega t)$$

参数值设为：$\gamma = 0.1$，$\omega = 1$，$\epsilon = 0.25$，$\Omega = 2$

初始条件设为：$x(0) = 1$，$\dot{x}(0) = 0$

分别画出 $\Gamma = 0.8$ 和 $\Gamma = 2.2$ 时的 $x-t$ 图和相图 $\dot{x}-x$。

提示：引入新变量 $y = \dot{x}$ 和 $z = \Omega t$，将上述运动方程化为三元一阶微分方程组。



```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Duffing(w,t,gamma,omiga,epsilou,Omega,Gamma): #定义函数
    x,y = w
    return np.array([y,Gamma*np.cos(Omega*t)-gamma*y+omiga*omiga*x-epsilou*x*x*x])#y=x. 一大串=x..
```

```python
t = np.arange(0,10,0.02)
y = odeint(Duffing,[1,0],t, args=(0.1,1,0.25,2,0.8))
plt.figure(figsize=(10, 8))
plt.plot(t,y[:,0])
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title('$\Gamma = 0.8$')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(y[:,0],y[:,1])
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title('$\Gamma = 0.8$')
```

```python
t = np.arange(0,10,0.02)
y = odeint(Duffing,[1,0],t, args=(0.1,1,0.25,2,2.2))
plt.figure(figsize=(10, 8))
plt.plot(t,y[:,0])
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title('$\Gamma = 2.2$')
```

```python
plt.figure(figsize=(10, 10))
plt.plot(y[:,0],y[:,1])
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title('$\Gamma = 2.2$')
```
